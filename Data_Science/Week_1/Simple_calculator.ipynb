## Simple CLI calculator repo

```python
def add(num1,num2): 
    return num1 + num2
```

```python
def subtract(num1,num2): 
    return num1 - num2
```

```python
def multiply(num1,num2): 
    return num1 * num2
```

```python
def divide(num1,num2): 
    if num2 == 0 : 
        return 'Error ! Division by zero '
    else : 
        return num1 / num2
```

```python
def mode(num1,num2): 
    return num1 % num2
```

```python
def show_calculator():
    print('Number one is = ' , num1)
    print('Number one is = ' , num2)
    print('-----------------------------')
    print(" ///  Simple CLI Calculator /// ")
    print('-----------------------------')
    print('Select Operation : ')
    print("+. Add")
    print("-. Subtract")
    print("*. Multiply")
    print("/. Divide")
    print("%. Mode")
    print('-----------------------------')

    choice = input('Enter The Operations : ')

    if choice == '+':
        print('{ + } : The Operations is add')
        print(f'{num1} + {num2}  = {add(num1,num2)}')
    elif choice == '-':
        print('{ - } : The Operations is Subtract')
        print(f'{num1} - {num2}  = {subtract(num1,num2)}')
    elif choice == '*':
        print('{ * } : The Operations is multiply')
        print(f'{num1} * {num2}  = {multiply(num1,num2)}')
    elif choice == '/':
        print('{ / } : The Operations is divide')
        print(f'{num1} / {num2}  = {divide(num1,num2)}')
    elif choice == '%':
        print('{ % } : The Operations is mode')
        print(f'{num1} % {num2}  = {mode(num1,num2)}')
    else:
        print("Invalid input. Please select a valid operation.")
```

```python
num1 = int(input('Enter The Number One : '))
print('Number one is = ' , num1)
num2 = int(input('Enter The Number Two : '))
print('Number one is = ' , num2)
```

```python
show_calculator()
```
